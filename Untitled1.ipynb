# Vectors

## What is a Vector?

You already know about vectors from Physics — a quantity that has both **direction** and **magnitude**. Like Displacement, Velocity, Acceleration...

In 2D (Two-Dimensional) space, it can be represented as:

$$\vec{A} = \begin{bmatrix} A_x \\ A_y \end{bmatrix}$$

where:

- $A_x$ = x-component
- $A_y$ = y-component

## But is it the Same as Vectors in CS/Math?

Before answering that, let's take a short journey into Dimensions.

### What is a Dimension?

In physics, a dimension is an independent direction in which movement or variation is possible. But we can simply think of it as the number of coordinates needed to describe a position.

**1-Dimensional (1D)**  
Only one coordinate is needed.

Examples:

* A number line
* Position on a straight road

Representation:
$$[x]$$

**2-Dimensional (2D)**  
Two coordinates are needed.

Examples:

* A sheet of paper
* A computer screen

Representation:
$$[(x, y)]$$

**3-Dimensional (3D)**  
Three coordinates are needed.

Examples:

* Physical space
* Real-world objects

Representation:
$$[(x, y, z)]$$

Example:
$$[(3, 4, 5)]$$

But what about **beyond 3 Dimensions**?  
Mathematics is not limited to physical space.  
A point can have any number of coordinates:

$$(x_1, x_2, x_3, \dots, x_n)$$

Example:

$$(20, 170, 65, 3.8)$$

This point exists in 4-dimensional space.

## Dimension in Machine Learning

In Machine Learning, each feature corresponds to one dimension.

Example:

A student described by:

* Age
* Height
* Weight

can be represented as:

$$[(20, 170, 65)]$$

This is a point in a 3-dimensional feature space.

If we add CGPA:

$$[(20, 170, 65, 3.8)]$$

the data point becomes 4-dimensional.

## Vectors in Mathematics and Computer Science

Now let's go back to our previous topic.

### Same Idea as Physics

A vector in mathematics is fundamentally the same concept as in physics.

In mathematics and computer science, we usually focus on the coordinates that describe the vector. For example:

$$\begin{bmatrix} 3 \\ 4 \end{bmatrix}$$

can be interpreted as:

* A displacement of 3 units along the x-axis and 4 units along the y-axis
* A point in a 2-dimensional space
* A data sample with two features

The underlying idea is the same.

### Why This Representation is Better

Drawing vectors works well in 2D and 3D. However, Machine Learning often deals with hundreds or thousands of dimensions.

For example, a student's data may be represented as:

$$[20, 170, 65, 3.8]$$

A modern dataset may contain:

$$[x_1, x_2, x_3, \dots, x_{1000}]$$

A 1000-dimensional vector cannot be visualized geometrically. Representing vectors as ordered lists of numbers allows us to work in any number of dimensions.

### A Vector as a Point

A vector can be viewed as a location in space.

Example:

$$\begin{bmatrix} 3 \\ 4 \end{bmatrix}$$

represents the point $(3, 4)$ in 2-dimensional space.

Similarly:

$$\begin{bmatrix} 2 \\ 5 \\ 7 \end{bmatrix}$$

represents a point in 3-dimensional space.

### Column Vectors

The standard representation in linear algebra is the column vector.

$$\mathbf{x} = \begin{bmatrix} 1 \\ 2 \\ 3 \end{bmatrix}$$

Dimension: $3 \times 1$, meaning:

* 3 rows
* 1 column

Most linear algebra formulas assume vectors are column vectors.

### Row Vectors

A vector can also be written horizontally.

$$\mathbf{x} = \begin{bmatrix} 1 & 2 & 3 \end{bmatrix}$$

Dimension: $1 \times 3$, meaning:

* 1 row
* 3 columns

This is called a row vector.

### Why Do We Need Both?

Many matrix operations require specific orientations.

For example:

$$\begin{bmatrix} 1 & 2 & 3 \end{bmatrix} \begin{bmatrix} 4 \\ 5 \\ 6 \end{bmatrix}$$

is valid matrix multiplication and produces a scalar.  
The first vector is a row vector.  
The second is a column vector.

So distinguishing between row and column vectors becomes important in linear algebra and machine learning.

# Vectors vs Sets

Vectors and sets may both contain multiple values, but they are fundamentally different.

## Order

A vector is ordered.

$$[1, 2, 3]$$

is different from

$$[3, 2, 1]$$

because the positions of the elements matter. In Machine Learning, each position usually corresponds to a specific feature. For example:

$$[20, 170, 65]$$

Changing the order changes the meaning.

A set is unordered.

$$\{1, 2, 3\}$$

and

$$\{3, 2, 1\}$$

represent the same set. The order of elements has no meaning.

## Duplicates

Vectors allow duplicate values.

$$[1, 1, 2, 2, 3]$$

is a valid vector. Each occurrence occupies its own position.

Sets do not allow duplicates.

$$\{1, 2, 3\}$$

Repeated elements are automatically ignored.

## Why Vectors Are Used in Machine Learning

Machine Learning data has:

* Ordered features
* Repeated values

For example:

$$[20, 20, 65, 65]$$

is perfectly valid data. Since order and duplicates both matter, vectors are a natural representation for data — sets are not.

# Vector Operations

Let:

$$\mathbf{a} = \begin{bmatrix} 1 \\ 2 \\ 3 \end{bmatrix}, \qquad \mathbf{b} = \begin{bmatrix} 4 \\ 5 \\ 6 \end{bmatrix}$$

### 1. Vector Addition

Add corresponding elements.

$$\mathbf{a} + \mathbf{b} = \begin{bmatrix} 1+4 \\ 2+5 \\ 3+6 \end{bmatrix} = \begin{bmatrix} 5 \\ 7 \\ 9 \end{bmatrix}$$

### 2. Scalar Multiplication

Multiply every element by a scalar.

$$2\mathbf{a} = 2 \begin{bmatrix} 1 \\ 2 \\ 3 \end{bmatrix} = \begin{bmatrix} 2 \\ 4 \\ 6 \end{bmatrix}$$

### 3. Hadamard Product

The Hadamard product is element-wise multiplication.

$$\mathbf{a} \odot \mathbf{b} = \begin{bmatrix} 1 \times 4 \\ 2 \times 5 \\ 3 \times 6 \end{bmatrix} = \begin{bmatrix} 4 \\ 10 \\ 18 \end{bmatrix}$$

The result is still a vector.

### 4. Dot Product

The dot product multiplies corresponding elements and then sums the results.

$$\mathbf{a} \cdot \mathbf{b} = 1 \times 4 + 2 \times 5 + 3 \times 6 = 4 + 10 + 18 = 32$$

Unlike the Hadamard product, the result is a scalar.

## A Note for Machine Learning

In mathematics and physics, you may have learned two common types of vector multiplication:

* Dot Product
* Cross Product

However, in Machine Learning, the Cross Product is rarely used. Instead, the two vector products you will encounter most often are:

* Hadamard Product (element-wise multiplication)
* Dot Product

### Mathematical Notation

Hadamard Product:

$$\mathbf{a} \odot \mathbf{b}$$

Dot Product:

$$\mathbf{a} \cdot \mathbf{b}$$

# Matrices

Now that we understand scalars, vectors, and tensors, let's focus on one of the most important mathematical tools in Machine Learning: the **matrix**.

## What is a Matrix?

A matrix is a **2-dimensional array** of numbers arranged in rows and columns.

### Example

$$A = \begin{bmatrix} 1 & 2 & 3 \\ 4 & 5 & 6 \\ 7 & 8 & 9 \end{bmatrix}$$

This matrix has 3 rows and 3 columns.

### Shape

$$( 3, 3 )$$

## Why Are Matrices Important in ML?

Most machine learning datasets are represented as matrices. For example:

$$X = \begin{bmatrix} 20 & 170 & 65 \\ 22 & 175 & 70 \\ 19 & 168 & 60 \end{bmatrix}$$

Here:
- Each row represents a data sample
- Each column represents a feature

| Age | Height | Weight |
|-----|--------|--------|
| 20  | 170    | 65     |
| 22  | 175    | 70     |
| 19  | 168    | 60     |

This entire dataset can be stored as a single matrix.

## Matrix Indexing (Mathematical Notation)

Consider the matrix:

$$A = \begin{bmatrix} 1 & 2 & 3 \\ 4 & 5 & 6 \\ 7 & 8 & 9 \end{bmatrix}$$

We denote its entries by $a_{ij}$ where:

- $i$ = row index
- $j$ = column index

### Entry

The element at row $i$ and column $j$ is called an **entry**.

Example:

$$a_{23} = 6$$

because the entry in the **2nd row** and **3rd column** is 6.

> Mathematical indexing starts from **1**, not 0.

### Row

The $i$-th row is written as $a_{i:}$.

For example, the second row is $[4, 5, 6]$.

In Machine Learning, a row usually represents a single data sample.

### Column

The $j$-th column is written as $a_{:j}$.

For example, the second column is:

$$\begin{bmatrix} 2 \\ 5 \\ 8 \end{bmatrix}$$

In Machine Learning, a column usually represents a feature.

# Matrix Operations

Let:

$$A = \begin{bmatrix} 1 & 2 \\ 3 & 4 \end{bmatrix}, \qquad B = \begin{bmatrix} 5 & 6 \\ 7 & 8 \end{bmatrix}$$

```python
import numpy as np

A = np.array([
    [1, 2],
    [3, 4]
])

B = np.array([
    [5, 6],
    [7, 8]
])
```

## 1. Matrix Addition

Matrix addition is performed by adding corresponding entries.

### Mathematically

$$A + B = \begin{bmatrix} 1+5 & 2+6 \\ 3+7 & 4+8 \end{bmatrix} = \begin{bmatrix} 6 & 8 \\ 10 & 12 \end{bmatrix}$$

### NumPy

```python
A + B
```

### Output

```python
array([
    [ 6,  8],
    [10, 12]
])
```

## 2. Scalar Multiplication

Multiply every entry of the matrix by a scalar.

### Mathematically

$$2A = \begin{bmatrix} 2 & 4 \\ 6 & 8 \end{bmatrix}$$

### NumPy

```python
2 * A
```

### Output

```python
array([
    [2, 4],
    [6, 8]
])
```

## 3. Hadamard Product

The Hadamard product is **element-wise multiplication**.

### Mathematically

$$A \odot B = \begin{bmatrix} 1 \times 5 & 2 \times 6 \\ 3 \times 7 & 4 \times 8 \end{bmatrix} = \begin{bmatrix} 5 & 12 \\ 21 & 32 \end{bmatrix}$$

### NumPy

```python
A * B
```

### Output

```python
array([
    [ 5, 12],
    [21, 32]
])
```

> Notice that `*` performs element-wise multiplication for NumPy arrays.

## 4. Matrix Multiplication

Unlike the Hadamard product, matrix multiplication combines rows and columns.

### Mathematically

$$AB = \begin{bmatrix} 1 & 2 \\ 3 & 4 \end{bmatrix} \begin{bmatrix} 5 & 6 \\ 7 & 8 \end{bmatrix} = \begin{bmatrix} 19 & 22 \\ 43 & 50 \end{bmatrix}$$

### NumPy

```python
A @ B
```

or

```python
np.matmul(A, B)
```

### Output

```python
array([
    [19, 22],
    [43, 50]
])
```

For matrix multiplication, each entry is obtained by taking the **dot product** of a row from the first matrix and a column from the second matrix.

For example:

$$(AB)_{11} = (1 \times 5) + (2 \times 7) = 19$$

## Example: CGPA Calculation

Suppose we have the grades of three students in three courses.

$$G = \begin{bmatrix} 3.75 & 3.50 & 4.00 \\ 3.25 & 3.75 & 3.50 \\ 4.00 & 3.75 & 3.75 \end{bmatrix}$$

Shape: $(3, 3)$

Rows represent students. Columns represent courses:

- Mathematics
- Physics
- Programming

Now suppose the credit hours are:

$$C = \begin{bmatrix} 3 \\ 3 \\ 4 \end{bmatrix}$$

Shape: $(3, 1)$

The weighted grade points are obtained by:

$$GC = \begin{bmatrix} 3.75 & 3.50 & 4.00 \\ 3.25 & 3.75 & 3.50 \\ 4.00 & 3.75 & 3.75 \end{bmatrix} \begin{bmatrix} 3 \\ 3 \\ 4 \end{bmatrix} = \begin{bmatrix} 37.75 \\ 35.50 \\ 38.25 \end{bmatrix}$$

Each entry represents the total weighted grade points of a student.

To get the CGPA, divide by the total credits:

$$3 + 3 + 4 = 10$$

$$\text{CGPA} = \begin{bmatrix} 3.775 \\ 3.550 \\ 3.825 \end{bmatrix}$$

### NumPy

```python
import numpy as np

G = np.array([
    [3.75, 3.50, 4.00],
    [3.25, 3.75, 3.50],
    [4.00, 3.75, 3.75]
])

C = np.array([
    [3],
    [3],
    [4]
])

weighted_points = G @ C

cgpa = weighted_points / C.sum()

print(cgpa)
```
